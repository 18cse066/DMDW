```
data.head()
```

```
data.dropna(axis=0,inplace=True) ##remove the null values first method rows when large data we are having
data.isnull().sum()
```

DATA VISUALIZATION

```
plt.figure(figsize=(10,8))
plt.scatter(data['t1'],data['t2'],c='green')
plt.xlabel('t1')
plt.ylabel('t2')
plt.show()
```

```
plt.figure(figsize=(15,8))
plt.hist(data['is_holiday'],edgecolor='white',bins=25)
plt.title('histogram of t1')
plt.xlabel('t1')
plt.ylabel('is_weekend')
plt.show()
```

```
sns.set(style='darkgrid')
sns.regplot(x=data['t1'],y=data['t2'])
```

```
sns.set(style='darkgrid')
sns.regplot(x=data['t1'],y=data['t2'],marker='*',fit_reg=False)
```

```
sns.distplot(data['t1'])
```

```
sns.distplot(data['t1'],kde=False)
```

```
sns.distplot(data['t1'],kde=False,bins=10)
```

```
plt.figure(figsize=(20,10))
sns.countplot(x=data['t1'],data=data)
```

```
plt.figure(figsize=(25,10))
sns.countplot(x=data['t1'],data=data,hue='t2')
```

```
sns.boxplot(y='t1',data=data)
```

```
plt.figure(figsize=(15,6))
sns.distplot(data.t1[data.t2==0])
sns.distplot(data.t1[data.t2==1])
plt.legend(['0','1'])
plt.show()
```
